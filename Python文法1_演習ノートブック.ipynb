## 4 算術演算子

**算術演算子**とは、数学で習った足し算や引き算など数値同士の計算を行うものです。
```python
x = 5 + 3
```
のように、数値の間におくことで計算ができます。  
足し算や引き算の演算子は、数学の記号と共通しますが、掛け算と割り算、べき乗は、特有の記号となります。掛け算は`*`（アスタリスク）記号、割り算は`/`（スラッシュ）記号、べき乗は2つの`*`記号で表します。また、余りのある割り算の商は`//`で、余りは`%`で計算できます。

|算術演算子|算術内容|
|:-|:-|
|+|足し算|
|-|引き算|
|<font color = "red">*</font>|掛け算|
|<font color = "red">/</font>|割り算|
|<font color = "red">//</font>|割り算(整数部)|
|<font color = "red">%</font>|割り算(余り)|
|<font color = "red">**</font>|べき乗|

以下のコードセルを実行してみましょう。

````python
# 足し算
1 + 1
````

````python
# 引き算
1 - 1
````

````python
# 掛け算（1×2）
1 * 2
````

````python
# 割り算 (1÷2)
1 / 2
````

````python
# 割り算 (整数部)
3 // 2
````

````python
# 割り算 (余り)
3 % 2
````

````python
# べき乗 (2の3乗)
2 ** 3
````

### Tips: 算術演算子の優先度

演算子が複数ある場合、優先度の高い演算子順に評価されます。  
算術演算子の優先度は、べき乗 > 掛け算・割り算 > 足し算・引き算の順となっており、数学のように足し算より掛け算の優先度が高く、`()`により優先度が変更できます。  
また、優先度が同じ場合には左から計算が行われます。  

*ちなみに、演算子には算術演算子の他に比較演算子、論理演算子などがあります。*

どのような答えになるかを予想し、下のコードセルを実行してみましょう。

````python
# 優先度：掛け算＞足し算
print(1 + 2 * 3)
````

````python
# 優先度：かっこ内の足し算＞掛け算
print((1 + 2) * 3)
````

````python
# 優先度：べき乗＞掛け算
print(2 * 2 ** 3)
````

 ### 型変換

データ型は型変換の関数によって変換することができます。
型変換の関数にはデータをint型に変換する`int()`やstr型に変換する`str()`があります。

**型変換の関数**

|関数|出力の型|
|:-|:-|
|int()|整数のint型へ|
|float()|小数のfloat型へ|
|str()|文字列のstr型へ|
|bool()|真偽値のbool型へ|

以下にいくつか例を載せています。実行し、結果を確認してください。

````python
#例1 str -> int
x = '100'
x = int(x)
type(x)
````

下のコードを実行してみてください。

````python
#例2 str -> intの失敗例
x = '10円'
x = int(x)
type(x)
````

float型をint型にするとき、小数点以下は切り捨てられます。  
下のコードを実行してみてください。

````python
#例3 float -> int
x = 100.6
x = int(x)
print(x)
print(type(x))
````

str型をfloat型に変更することもできます。  
intに変換する場合と同じように、数字以外を含む文字列だとエラーが発生するので気をつけてください。

````python
#例4 str -> float
x = '100.5'
x = float(x)
print(x)
print(type(x))
````

以下はint型、float型のデータをstr型に変換するコードです。

````python
#例5 int, float -> str
x = 100
y = 1.5
x = str(x)
y = str(y)
print(x)
print(y)
print(type(x))
print(type(y))
````

bool型はint型、float型との間で変換を行う場合、`True`が**1**、`False`が`0`として扱われます。  
下のコードを実行してみてください。

````python
#例6 bool -> int と　float -> bool
x = True
y = 0.0
x = int(x)
y = bool(y)
print(x)
print(y)
print(type(x))
print(type(y))
````
